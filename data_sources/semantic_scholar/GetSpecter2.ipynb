```python
import pandas as pd
import urllib, urllib.request
import requests
import time, re
import json, glob
from difflib import SequenceMatcher
from ast import literal_eval
from collections import Counter
from pprint import pprint
```

```python
with open('apikey.txt', encoding = 'utf-8') as f:
    apikey = f.read().strip()
```

```python
def get_embedding(paperId):
    
    url = 'https://api.semanticscholar.org/graph/v1/paper/' + paperId 
    
    data = 'not found'
    
    response = requests.get(url, 
                            headers={'X-API-KEY': apikey},
                           params={'fields': 'embedding.specter_v2'})
    if response.status_code == 200:
        jsonobj = response.json()
        if 'embedding' in jsonobj:
            embedobj = jsonobj['embedding']
            if embedobj is not None and 'vector' in embedobj:
                data = embedobj['vector']
            
        else:
            data = 'no embedding'
    
    return data 
                
```

```python
meta = pd.read_csv('../../metadata/litstudies/LitMetadataWithS2.tsv', sep = '\t')
meta.head()
```

```python
titles = []
embeddings = []

for idx, row in meta.iterrows():
    if isinstance(row['paperId'], str):
        embedding = get_embedding(row['paperId'])
        if not isinstance(embedding, str):
            titles.append(row['title'])
            embeddings.append(embedding)
        
            if len(titles) % 10 == 1:
                print(len(titles))
        else:
            print(embedding)
    
    if len(titles) > 200:
        break
```
