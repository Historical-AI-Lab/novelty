# Prolific Critics

```python
import pandas as pd
from ast import literal_eval
from collections import Counter
```

```python
meta = pd.read_csv('../metadata/litstudies/LitMetadataWithS2.tsv', sep = '\t')
meta.head()
```

```python
authors = Counter()
author_papers = dict()
meta.authors = meta.authors.apply(literal_eval)
```

```python
for idx, row in meta.iterrows():
    authlist = row['authors']
    paperId = row['paperId']
    for auth in authlist:
        authors[auth] += 1
        if auth not in author_papers:
            author_papers[auth] = []
        if not pd.isnull(paperId):
            author_papers[auth].append(paperId)
```

```python
for auth, count in authors.items():
    authors[auth] = len(author_papers[auth])

authors.pop('anonymous', None)
```

```python
with open('prolific_critics.tsv', 'w') as f:
    f.write('author\tpapercount\tbirthyear\tpapers\n')
    for auth, count in authors.most_common(100):
        f.write(f'{auth}\t{len(author_papers[auth])}\t\t{str(author_papers[auth])}\n')
```
