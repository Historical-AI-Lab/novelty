### Function to get the last names and make sure every name is there

```python
sample1 = "['Lawrence G. Derthick', 'Brice Harris', 'Norma V. Fornaciari', 'Nicholson B. Adams', 'Arthur S. Adams', 'Frederick Burkhardt', 'Lander MacClintock', 'Henry M. Hoenigswald', 'Alfred Senn', 'Henri Peyre', 'Blyden Jackson', 'Henry W. Sams', 'Zbigniew Folejewski', 'James Gray', 'Willard Thorp']"
sample2 = "['Bruce Mitchell']"

def getLastName(singleName):
    decomposedName = singleName.split()
    return(decomposedName[-1])

def makeLastNames(author_names):
    bracketsRemoved= author_names[1:-1]
    if "," in bracketsRemoved:
        splitNames = bracketsRemoved.split(",") 
        cleanedNames = [i[1:-1] for i in splitNames]
        lastNames = [getLastName(i) for i in cleanedNames]
        return(", ".join(lastNames))
    else:
        name = bracketsRemoved[1:-1]
        lastName = getLastName(name)
        return(lastName)

print(makeLastNames(sample1))
print(makeLastNames(sample2))
```
